```python
##Create A Black Image
img=np.zeros((512,512,3),np.uint8)
cv.rectangle(img,(100,100),(300,300),(0,255,0),3)
plt.imshow(img)
```

```python
cv.circle(img,(447,63),63,(0,0,255),-1)
plt.imshow(img)
```

```python
cv.ellipse(img,(50,50),(100,100),0,0,180,255,-1)
plt.imshow(img)
```

```python
img = np.zeros((600,600,3), np.uint8)
img=cv.line(img,(100,100),(300,300),(255,0,0),5)
img=cv.rectangle(img,(100,100),(400,500),(255,0,0),5)
img=cv.circle(img,(300,200),60,(255,0,0),5)
img=cv.ellipse(img,(250,350),(100,150),0,0,180,255,-1)
pts = np.array([[250,350],[210,310],[260,330],[240,315]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img,[pts],True,(250,0,0))
plt.imshow(img)
```
