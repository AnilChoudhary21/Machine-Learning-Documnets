### Stemming : Removing the Ending parts of the words

```python
don='Programmer,Programs,Programee'
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import PorterStemmer
ps=PorterStemmer()
word=word_tokenize(don)
```

```python
for i in word:
     print(i, ps.stem(i))
```

### Lemitization : A Tree of how words are linked with each others and for every word it maps with the root nodes

```python
import nltk
nltk.download('wordnet')

words=['rocks','corpora','better']

from nltk.stem import WordNetLemmatizer
wlm=WordNetLemmatizer()

for i in words:
    print(i,wlm.lemmatize(i,pos ='a'))
```

```python
df=pd.read_csv(r'D:\Data Science\Dataset\Amazon_Unlocked_Mobile\Amazon_Unlocked_Mobile.csv')
```

```python
df.head()
```

```python
x=df['Reviews']
y=df['Rating']
```

```python
np.unique(y.values)
```

```python
sentiment=[]
for i in y.values:
    if i<3:
        sentiment.append('negative')
    elif i==3:
        sentiment.append('Neutral')
    else:
        sentiment.append('positive')
        
```

```python
df['sentiment']=sentiment
```

```python
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
d1=df[df['sentiment']=='Neutral']
d2=df[df['sentiment']=='negative'].sample(31765)
d3=df[df['sentiment']=='positive'].sample(31765)
```

```python
m1=d1.append(d2)
final_dataset=m1.append(d3)
```

```python
final_dataset['sentiment'].value_counts()
```

```python
y=final_dataset['sentiment']
x=final_dataset['Reviews']
```

```python
final_dataset.dropna(how='any',axis=0,inplace=True)
```

```python
x=df['Reviews']
y=df['sentiment']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(stop_words='english',ngram_range=(1,2))
tf.fit(X_train)
X_train_transformed=tf.transform(X_train)
X_test_transformed=tf.transform(X_test)
```

```python
X_train_transformed
```

```python
tf.get_feature_names()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words =set(stopwords.words("english"))

print(len(stop_words))

stop_words=list(stop_words)
```

After Preprocessing

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(X_train_transformed,y_train)
dt.predict(X_test_transformed)
print(confusion_matrix(y_test,dt.predict(X_test_transformed)))
```

```python
print(accuracy_score(y_test,dt.predict(X_test_transformed)))
```
