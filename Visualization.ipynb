```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\Data Science\Dataset\Rossmann Store Sales\train.csv')
```

```python
df['Date'].nunique()
```

```python
df_date=pd.DataFrame(df.groupby('Date')['Sales'].sum())
df_date.shape
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_date[0:50])
plt.xlabel('Sales')
plt.xticks(rotation=45)
plt.ylabel('Date')
plt.title('Daily Sales Forecast')
```

```python
df_date_promo=pd.DataFrame(df[df['Promo']==1].groupby('Date').sum()['Sales'])
df_date_No_promo=pd.DataFrame(df[df['Promo']==0].groupby('Date').sum()['Sales'])
```

```python
plt.plot(df_date_promo['Sales'].values[0:20],color='r')
plt.plot(df_date_No_promo['Sales'].values[0:20],color='b')
```
